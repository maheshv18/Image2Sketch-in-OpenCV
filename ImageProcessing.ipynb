```python
import pandas as pd
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img=cv2.imread('noddy.jpg')
```

```python
cv2.imshow("img",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
def show_image(img_arr):
    rgb_arr=cv2.cvtColor(img_arr,cv2.COLOR_BGR2RGB)
    plt.imshow(rgb_arr)
    plt.axis(False)
```

```python
show_image(img)
```

```python
gray_img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
show_image(gray_img)
```

```python
show_image(255-gray_img)
inverted_img=255-gray_img
```

```python
blur_img=cv2.GaussianBlur(inverted_img,(111,111),0)
```

```python
show_image(blur_img)
```

```python
sketch=cv2.divide(blur_img,inverted_img,scale=256.0)
```

```python
show_image(sketch)
```
